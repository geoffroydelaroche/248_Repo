```python
df=pd.read_csv('https://ironhack.school/asset-v1:IRONHACK+DAFT+201910_PAR+type@asset+block@IMDB-Movie-Data.csv')
df.head()
```

```python
agg_df=df[['Year','Rating']].groupby('Year').mean().copy()
agg_df
```

```python
agg_df.reset_index(inplace=True)
agg_df
```

```python
sns.lineplot(x='Year', y='Rating', data=agg_df)
```

```python
agg_df2=df[['Year','Revenue (Millions)']].groupby('Year').mean().copy()
```

```python
agg_df2.reset_index(inplace=True)
agg_df2
```

```python
sns.scatterplot('Year','Revenue (Millions)', data=agg_df2)
```

```python
sns.scatterplot('Year','Revenue (Millions)', data=df)
```

```python
sns.lmplot('Year','Revenue (Millions)', data=df)
```
