## 2.7 Binning

There are 3 main ways of binning:
    * equal width bins
    * equal frequency bins
    * custom ones

### 2.7.1 Equal width bins

```python
labels=['Terrible','Bas','ok','Good','Amazing']
```

```python
df.IMDb.describe()
```

```python
pd.cut(df.IMDb, 5)
```

```python
pd.cut(df.IMDb, 5, labels=labels)
```

```python
bins=pd.cut(df.IMDb, 5, labels=labels)
```

```python
bins.cat.codes
```

```python
bins, representation = pd.cut(df.IMDb, 5, labels=labels, retbins=True)
```

```python
bins
```

```python
representation
```

### 2.7.2 Equal frequency bins

```python
bins=pd.qcut(df.IMDb, 5, labels=labels)
```

```python
bins
```

```python
bins, representation = pd.qcut(df.IMDb, 5, labels=labels, retbins=True)
```

```python
bins
```

```python
representation
```

```python
bins.value_counts()
```
