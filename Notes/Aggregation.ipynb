## 3.2 filter()

```python
test=df.groupby('Age').IMDb.mean()
test
```

```python
test[test>7.2]
```

```python
age_group_selected=test[test>7.2].index
age_group_selected
```

```python
df.Age.isin(age_group_selected)
```

```python
df[df.Age.isin(age_group_selected)]
```

```python
df.groupby('Age').filter(lambda x: x.IMDb.mean()>7.2)
```

## 3.3 transform()

```python
test=df.groupby('Age').IMDb.mean()
test
```

```python
def compare(x):
    if isinstance (x.Age, str): #np.nan is float
        return x.IMDb-test[x.Age]
    else:
        return np.nan
```

```python
df.apply(compare, axis=1) #apply the function to every single row of dataframe
```

Function transform can't handle missing values, therefore we need to fillna

```python
df2=df[['Age','IMDb']].copy()
df2
```

```python
df2.Age.fillna('missing',inplace=True)
df2
```

```python
df2.groupby('Age').IMDb.transform(lambda x: x-x.mean())
```

Ranking per group

```python
test=df2[['Age','IMDb']].groupby('Age') #split my df into separate dataframes with different age category in each

for i,j in test:
    print(i)
```

```python
new_series=pd.Series()
```

```python
for i, j in test: #loop over groupby object. On every iteration i is age category, j is dataframe with i
    new_series=new_series.append(j.IMDb.rank(ascending=False)) #get rank for each group separately and append into new_series

new_series.sort_index()
```

```python
df2.groupby('Age').IMDb.transform(lambda x: x.rank(ascending=False))
```

```python
# top 1 in each category
df.loc[df2.groupby('Age').IMDb.transform(lambda x: x.rank(ascending=False)).sort_values().head(12).index]
```
