# Silhouette and Davies Bouldin

```python
from sklearn.datasets import make_blobs, make_moons
```

```python
X, y=make_blobs(random_state=42)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
single=AgglomerativeClustering(n_clusters=3, linkage='single')
kmeans=KMeans(3)
```

```python
single.fit(X)
kmeans.fit(X)
```

```python
ypred1=single.labels_
ypred1
```

```python
ypred2=kmeans.labels_
ypred2
```

```python
_,ax=plt.subplots(1,2,figsize=(12,6))
ax[0].scatter(X[:,0], X[:,1], c=ypred1)
ax[1].scatter(X[:,0], X[:,1], c=ypred2)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
print('Model 1. Silhouette score is', silhouette_score(X, ypred1))
print('Model 2. Silhouette score is', silhouette_score(X, ypred2))
print('')
print('Model 1. Bouldin score is', davies_bouldin_score(X, ypred1))
print('Model 2. Bouldin score is', davies_bouldin_score(X, ypred2))
```

# Keyword selection

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
categories=['comp.graphics','talk.politics.guns', 'sci.space','rec.sport.hockey']
```

```python
dataset=fetch_20newsgroups(subset='all', categories=categories, shuffle=True, random_state=11)
```

```python
len(dataset)
```

```python
len(dataset.data)
```

```python
dataset.target_names
```

```python
print(dataset.data[0])
```

```python
print(dataset.data[100])
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer
```

```python
vector=TfidfVectorizer(max_df=0.5, max_features=1000, stop_words='english')
```

```python
X=vector.fit_transform(dataset.data)
```

```python
X
```

```python
km=KMeans(n_clusters=len(dataset.target_names), max_iter=1000, n_jobs=-1)
km.fit(X)
```

```python
silhouette_score(X, km.labels_)
```

```python
davies_bouldin_score(X.toarray(), km.labels_)
```

```python
terms=vector.get_feature_names()
```

```python
for i in range(len(dataset.target_names)):
    print(f'Cluster {i+1}:')
    for ind in km.cluster_centers_.argsort()[:,::-1][i,:10]:
        print(terms[ind], end=' ')
    print()
```
