## Let's start with a simple link https://api.github.com/

```python
import requests
import json
import pandas as pd
```

```python
url = 'https://api.github.com/'
response = requests.get(url) # We are trying to enter the URL and read the data
results = response.json() #We know that the response is JSON, so we say response.json() else we say response.content()
results
```

```python
# Our results is a dictionary. Let's convert it to a DataFrame
data = pd.DataFrame(results, index=[0]).T
data
```

```python
## Alternate method
data = pd.DataFrame(results.items())
data = data.set_index(0)
data
```

## Another example: More complex

```python
url = 'https://api.github.com/events'
response1 = requests.get(url) # We are trying to enter the URL and read the data
results1 = response1.json() #We know that the response is JSON, so we say response.json() else we say response.content()
results1
```

```python
data1 = pd.DataFrame(results1)
data1.head()
```

```python
data1['actor'][0] #There are some dictionaries as elements
```

```python
flattened_data = pd.DataFrame(dict(data1['actor'])).T #The dict function permit to have the index
flattened_data
```

```python
data1.shape
```

```python
flattened_data.shape
```

```python
pd.concat([data1, flattened_data], axis=1) #In order to concatenate the data1 with flattened_data, we specified axis=1 in order to concatenate as columns
```

```python
to_flatten= ['actor','repo','payload','org']  #Represent every columns with a dictionnary in it. We have to seperate those columns into several columns for each

for i in to_flatten:
    flattened_data = pd.DataFrame(dict(data1[i])).T  # Create the sub DataFrame of the column
    
    flattened_data.columns=[f'{i}_{col}' for col in flattened_data.columns]  #In order to rename the column with a reminder of the original column
    
    data1 = pd.concat([data1, flattened_data], axis=1)  # Concat the DataFrame with the sub DataFrame
    
    data1 = data1.drop(columns=i) # In order to drop the original column which contain dictionnaries
data1
```

```python
# Other option: using json normalize 
url = 'https://api.github.com/events'
response2 = requests.get(url) 
results2 = response2.json() 
data2 = pd.DataFrame(results2)
data2.head()
```
