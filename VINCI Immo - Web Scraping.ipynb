```python
lst_url=[] # url list for all main pages (here we need to know how many main pages)
lst_nid=[] # list of nid code in order to get all projects details
lst_url_projets=[] # url list for all projects pages

import time

for i in range(3):
    url_progs=f'https://www.vinci-immobilier.com/api/offres/recherche/programme?medias[]=Site%20R%C3%A9sidentiel%20%2B%20Extranet%20client&phases_commerciales[]=Avant-premi%C3%A8re&phases_commerciales[]=Commercialisation&phases_commerciales[]=Lancement&phases_commerciales[]=Parkings%20seulement&site=v4&page={i}&items_per_page=11&region=11'
    lst_url.append(url_progs)

lst_url
```

```python
lst=[]
#get all projects nid & details from the main pages
for i in lst_url:
    data=r.get(i, headers=headers)
    data=data.json()
    listing_site = pd.DataFrame(data['results'])
    listing_site = listing_site[['nid','title','locality','postal_code','field_ref_operation']]
    lst.append(listing_site)
    time.sleep(7)
db = pd.concat(lst, axis=0, ignore_index=True)
db
```

```python
for i in db['nid']:
    lst_nid.append(i)
    
for i in lst_nid:
    url_proj=f'https://www.vinci-immobilier.com/api/offres/programme/{i}/lotsV4?site=v4&statut[]=Libre&statut[]=Optionn%C3%A9'
    lst_url_projets.append(url_proj)
    
lst_url_projets
```

```python
flats=[]
#get all flats details from the main pages
for ix,i in enumerate(lst_url_projets):
    data_f = r.get(i, headers=headers)
    data_f = data_f.json()
    listing_f = pd.DataFrame(data_f)
    listing_f = listing_f[['titre','nid','lots']]                    
    flats.append(listing_f)
    if ix%5==0:time.sleep(8)
```

```python
db_f = pd.concat(flats)
db_f
```
